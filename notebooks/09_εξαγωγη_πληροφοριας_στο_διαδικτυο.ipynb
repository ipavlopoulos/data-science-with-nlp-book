### Εισαγωγή από το βιβλίο

Στο προηγούμενο κεφάλαιο είδαμε πώς να επικοινωνούμε με τον Ιστό και πώς αποκτούμε πρόσβαση στα δεδομένα του. Είδαμε, δηλαδή, πώς να στέλνουμε αιτήματα, να κατανοούμε URLs και να λαμβάνουμε το περιεχόμενο μιας ιστοσελίδας με HTML μορφή. Σε αυτό το κεφάλαιο προχωράμε ένα βήμα παραπέρα και ασχολούμαστε με το πρόβλημα του πώς μετατρέπουμε το περιεχόμενο αυτό σε χρήσιμη πληροφορία.

Η γλώσσα HTML δεν είναι σχεδιασμένη για άμεση ανάλυση από ανθρώπους ή αλγορίθμους. Είναι μια γλώσσα που περιγράφει τη δομή και την παρουσίαση μιας σελίδας, όχι τη σημασιολογία των δεδομένων της. Μέσα σε ένα αρχείο HTML συνυπάρχουν διαφορετικά είδη περιεχομένου, όπως τίτλοι, παράγραφοι, σύνδεσμοι, μενού πλοήγησης, διαφημίσεις, χωρίς πάντα σαφή διάκριση για το τι είναι σημαντικό και τι όχι.

Η εξαγωγή πληροφορίας από τον Ιστό βοηθάει ακριβώς εκεί, στο να εντοπίσουμε, να απομονώσουμε και να δομήσουμε τα δεδομένα που μας ενδιαφέρουν. Για να το πετύχουμε αυτό, αξιοποιούμε τη δομή του HTML και αναπαριστούμε ως DOM. Αντί να αντιμετωπίζουμε τη σελίδα ως απλό κείμενο, τη βλέπουμε ως ένα δέντρο από στοιχεία, στο οποίο μπορούμε να πλοηγηθούμε, να αναζητήσουμε πρότυπα και να εξάγουμε συγκεκριμένα τμήματα.

```python
import requests
from bs4 import BeautifulSoup

header = {"User-Agent": "CrawlBase/1.0"}
url = "https://en.wikipedia.org/wiki/Data_science"
response = requests.get(url, headers = header)
soup = BeautifulSoup(response.text, "html.parser")
links = soup.find_all("a", href=True)
max_pages_visited = 10
for link in links:
    next_url = link.get("href")
    if next_url.startswith("http"):
      print("Visiting:", next_url)
      max_pages_visited-=1
      if max_pages_visited == 0:
        break
      try:
          r = requests.get(next_url, headers=header)
          print("  Status:", r.status_code)
      except:
          print("  Failed")
```

```python
response = requests.get(url, headers = header)
soup = BeautifulSoup(response.text, "html.parser")
links = soup.find_all("a", href=True)
max_pages_visited = 10
memory = set()
for link in links:
    next_url = link.get("href")
    if next_url in memory:
      print('Visited already: ', next_url)
      continue
    else:
      memory.add(next_url)
      if next_url.startswith("http"):
        print("Visiting:", next_url)
        max_pages_visited-=1
        if max_pages_visited == 0:
          break
        try:
            r = requests.get(next_url, headers=header)
            print("  Status:", r.status_code)
        except:
            print("  Failed")
```

### 9.4 Εφαρμογή: Εξαγωγή σχολίων (warmane)

```python
import requests
from bs4 import BeautifulSoup

url = "https://forum.warmane.com/forumdisplay.php?f=20&page=1"

try:
    # 1. HTTP Request
    req = requests.get(url, headers={'User-Agent': 'ScrapRE/5.0'})

    # 2. Συντακτική ανάλυση με BeautifulSoup
    soup = BeautifulSoup(req.text, "html.parser")

    # 3. Εξαγωγή Δεδομένων: Ψάχνουμε όλα τα links που είναι τίτλοι θεμάτων
    threads = soup.find_all("a", class_="title")

    print(f"Βρέθηκαν {len(threads)} θέματα:\n")

    for thread in threads:
        title = thread.text.strip() # Παίρνουμε το κείμενο
        link = "https://forum.warmane.com/" + thread['href'] # Παίρνουμε το URL
        print(f"Τίτλος: {title}")
        print(f"Σύνδεσμος: {link}\n" + "-"*20)

except Exception as e:
    print(f"Προέκυψε σφάλμα: {e}")
```

```python
def scrape_comments(url, headers={'User-Agent': 'ScrapRE/5.0'}):
    try:
        req = requests.get(url, headers=headers)
        soup = BeautifulSoup(req.text, "html.parser")

        # Στον Warmane, κάθε post βρίσκεται συνήθως σε ένα li με id "post_..."
        posts = soup.find_all("li", class_="postbitlegacy")

        print(f"--- Συλλογή σχολίων από: {url} ---\n")

        for post in posts:
            # 1. Εύρεση Συγγραφέα
            author = post.find("a", class_="username").text.strip()

            # 2. Εύρεση Ημερομηνίας
            date = post.find("span", class_="date").text.strip()

            # 3. Εύρεση Περιεχομένου (το κυρίως σχόλιο)
            comment_body = post.find("blockquote", class_="postcontent").text.strip()

            print(f"Χρήστης: {author} ({date})")
            print(f"Σχόλιο: {comment_body[:200]}...") # Τυπώνουμε τους πρώτους 200 χαρακτήρες
            print("-" * 30)

    except Exception as e:
        print(f"Σφάλμα κατά την ανάγνωση: {e}")

scrape_comments('https://forum.warmane.com/showthread.php?t=445626&s=70880da16ab90bd5c4d989f53b78e5f0')
```
