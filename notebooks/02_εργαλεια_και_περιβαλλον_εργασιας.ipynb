### Εισαγωγή από το βιβλίο

Η Επιστήμη Δεδομένων δεν αφορά μόνο αλγορίθμους και μαθηματικά μοντέλα. Στον πυρήνα της, είναι μια πρακτική δραστηριότητα που βασίζεται σε εργαλεία, περιβάλλοντα εργασίας και ροές ανάλυσης δεδομένων. Πριν ο αναγνώστης ασχοληθεί με αναπαραστάσεις, οπτικοποίηση, μηχανική μάθηση ή αξιολόγηση μοντέλων, είναι απαραίτητο να αποκτήσει μια στέρεη κατανόηση του οικοσυστήματος μέσα στο οποίο πραγματοποιείται η ανάλυση.

Στο κεφάλαιο αυτό παρουσιάζονται τα βασικά εργαλεία που χρησιμοποιούνται σήμερα στην Επιστήμη Δεδομένων, με έμφαση στην Python και στις βιβλιοθήκες που τη συνοδεύουν. Η Python έχει καθιερωθεί ως η κυρίαρχη γλώσσα του πεδίου, όχι μόνο λόγω της απλότητας και της εκφραστικότητάς της, αλλά και επειδή επιτρέπει την ομαλή μετάβαση από τον πειραματισμό στην παραγωγική ανάπτυξη συστημάτων. Παράλληλα, υποστηρίζει μια ενοποιημένη ροή εργασίας που καλύπτει όλα τα στάδια της ανάλυσης: από τη λήψη και τον καθαρισμό των δεδομένων έως την οπτικοποίηση και την εξαγωγή συμπερασμάτων.

Ιδιαίτερη έμφαση δίνεται στα Jupyter Notebooks και στα σύγχρονα περιβάλλοντα ανάπτυξης (IDEs), τα οποία αποτελούν βασικά εργαλεία τόσο για την εξερευνητική ανάλυση όσο και για την τεκμηρίωση της επιστημονικής σκέψης. Ο αναγνώστης θα δει πώς ο συνδυασμός κώδικα, σχολίων και γραφημάτων διευκολύνει την κατανόηση των δεδομένων και την αναπαραγωγιμότητα των πειραμάτων, αλλά και ποιες καλές πρακτικές πρέπει να ακολουθούνται ώστε ο κώδικας να παραμένει καθαρός, επεκτάσιμος και αξιόπιστος.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

messages = [
    "Μου αρέσει πολύ αυτό το προϊόν!",
    "Δεν μου άρεσε καθόλου η εξυπηρέτηση.",
    "Καλή εμπειρία συνολικά.",
    "Απογοητεύτηκα, ήταν ακριβό.",
    "Το συνιστώ ανεπιφύλακτα!"
]
```

```python
df = pd.DataFrame({
    "message": messages
})

# Προσθήκη μήκους μηνύματος
df["length"] = df["message"].apply(len)
```

```python
df.describe()
```

```python
plt.bar(df.index, df["length"])
plt.xticks(df.index, df["message"], rotation=45, ha="right")
plt.ylabel("Μήκος μηνύματος")
plt.title("Μήκος μηνυμάτων στα ελληνικά κοινωνικά δίκτυα")
plt.tight_layout()
plt.show()
```
