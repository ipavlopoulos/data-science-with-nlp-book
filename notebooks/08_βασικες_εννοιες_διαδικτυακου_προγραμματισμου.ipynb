### Εισαγωγή από το βιβλίο

Ο Παγκόσμιος Ιστός (Web) αποτελεί σήμερα μία από τις σημαντικότερες πηγές δεδομένων. Από ειδησεογραφικά άρθρα και επιστημονικές δημοσιεύσεις μέχρι κοινωνικά δίκτυα και κυβερνητικά δεδομένα, τεράστιες ποσότητες πληροφορίας είναι διαθέσιμες διαδικτυακά. Για έναν επιστήμονα δεδομένων, η δυνατότητα αυτόματης συλλογής και επεξεργασίας αυτών των δεδομένων είναι σχεδόν απαραίτητη.

Για να κατανοήσουμε πώς λειτουργεί ο ίδιος ο Ιστός πρέπει να μπορούμε να απαντήσουμε βασικές ερωτήσεις: Τι σημαίνει “κατεβάζω μία σελίδα”; Τι μορφή έχουν τα δεδομένα που λαμβάνω; Πώς επικοινωνεί ο υπολογιστής με έναν απομακρυσμένο εξυπηρετητή;

Το κεφάλαιο αυτό εισάγει τις βασικές έννοιες του διαδικτυακού προγραμματισμού που απαιτούνται για να απαντήσουμε στα παραπάνω ερωτήματα.


```python
import requests
response = requests.get("https://en.wikipedia.org/wiki/Data_science")
print(response.status_code)
```

```python
import requests

headers = {"User-Agent": "ResearchBot/1.0"}
response = requests.get("https://en.wikipedia.org/wiki/Data_science", headers=headers)

print(response.status_code)
```

```python
print(response.text[:500])
```

```python
print(response.headers)
```

```python
import requests
response = requests.get("https://example.com/nonexistent/", verify=False)
print(response.status_code)
```

```python
base_url = "https://example.com/search"

for page in range(1, 6):
    url = f"{base_url}?q=data+science&page={page}"
    print(url)
```

```python
import requests

params = {
    "q": "data science",
    "page": 2
}

response = requests.get("https://example.com/search",
                        params=params,
                        verify=False)

print(response.url)
```

```python
import requests

data = {
    "username": "john",
    "password": "1234"
}

response = requests.post("https://example.com/login", data=data, verify=False)
print(response.status_code)
print(response.text)
```
