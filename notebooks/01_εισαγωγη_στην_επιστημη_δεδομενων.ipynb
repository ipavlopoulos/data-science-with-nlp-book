### Εισαγωγή από το βιβλίο

Η επιστήμη δεδομένων αποτελεί ένα διεπιστημονικό πεδίο που συνδυάζει μεθόδους της πληροφορικής, της στατιστικής και της εφαρμοσμένης μαθηματικής μοντελοποίησης με στόχο την εξαγωγή γνώσης από δεδομένα. Στον πυρήνα της βρίσκεται η συστηματική μελέτη πραγματικών δεδομένων, τα οποία συχνά είναι μεγάλα σε όγκο, ετερογενή ως προς τη μορφή τους και θορυβώδη. Σε αντίθεση με την παραδοσιακή ανάλυση δεδομένων, η επιστήμη δεδομένων δεν περιορίζεται στην περιγραφή ή τη στατιστική σύνοψη, αλλά στοχεύει στην κατανόηση, πρόβλεψη και υποστήριξη λήψης αποφάσεων, αξιοποιώντας αλγορίθμους μηχανικής μάθησης και υπολογιστικές τεχνικές μεγάλης κλίμακας.

```python
texts = [
    "η επιστήμη δεδομένων",
    "δεδομένα και γλώσσα",
    "ανάλυση φυσικής γλώσσας"
]

# Λεξιλόγιο
vocabulary = sorted(set(" ".join(texts).split()))
vocabulary
```

```python
# Bag-of-Words αναπαράσταση
def vectorize(text, vocab):
    return [text.split().count(word) for word in vocab]

vectors = [vectorize(t, vocabulary) for t in texts]
vectors
```
